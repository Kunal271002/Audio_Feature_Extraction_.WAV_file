```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load an audio file (replace 'your_audio_file.wav' with your file path)
audio_file_path = './s1/an1.wav'
y, sr = librosa.load(audio_file_path)

# Compute MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # You can choose the number of MFCC coefficients (n_mfcc)

# Display MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(mfccs, ref=np.max), y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.show()
```

```python
from moviepy.editor import VideoFileClip
from pydub import AudioSegment
import os

# Directory containing the .avi video files
video_directory = '.\Dataset 1\s8\s8\s83'

# Directory to save the .wav audio files
audio_directory = '.\s8\s83'

# Create the output audio directory if it doesn't exist
os.makedirs(audio_directory, exist_ok=True)

# Loop through all .avi files in the video directory
for filename in os.listdir(video_directory):
    if filename.endswith(".avi"):
        # Construct the full paths
        video_path = os.path.join(video_directory, filename)
        audio_path = os.path.join(audio_directory, os.path.splitext(filename)[0] + ".wav")
        
        # Load the video
        video = VideoFileClip(video_path)
        
        # Extract the audio from the video
        audio = video.audio
        
        # Save the audio as a .wav file
        audio.write_audiofile(audio_path, codec='pcm_s16le')  # Use 'pcm_s16le' codec for WAV
        
        # Close the video and audio objects
        video.close()
        audio.close()

print("Conversion complete.")
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os

data_dir = './Dataset 2/s1'

def compute_mfcc(audio_file_path):
    y, sr = librosa.load(audio_file_path)
    
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    
    return mfccs

for filename in os.listdir(data_dir):
    if filename.endswith('.wav'):
        audio_file_path = os.path.join(data_dir, filename)
        
        mfccs = compute_mfcc(audio_file_path)
        
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(librosa.power_to_db(mfccs, ref=np.max), y_axis='mel', x_axis='time')
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'MFCC for {filename}')
        plt.show()
        
```
